## ML with Load_Boston()
#### 1. Import python packages

import numpy as np 
import pandas as pd 
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression

#### 2. Load data Boston

```python
dataBoston = load_boston()

df = pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston['feature_names']
)
df['PRICE'] = dataBoston['target']
df
```

#### 3. Correlation Plot

```python
corr = df.corr()
ticks = np.arange(14)
plt.imshow(corr, cmap = 'BuPu_r')
plt.colorbar()
plt.xticks(ticks, df.columns, rotation=60)
plt.yticks(ticks, df.columns)
plt.show()
```

#### 4. ML Model - Regression

```python
model = LinearRegression()
model.fit(df[['ZN', 'CHAS', 'RM', 'DIS', 'B']], df['PRICE'])
model.predict(df.head(1)[['ZN', 'CHAS', 'RM', 'DIS', 'B']])
```

```python
model = LinearRegression()
model.fit(df[['CRIM']], df['PRICE'])
model.predict(df.head(1)[['CRIM']])
```
